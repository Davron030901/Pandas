````python
import numpy as np
import pandas as pd
````

````python
sales_data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Sales': [150, 200, 180, 220, 190, 210, 165, 175],
    'Profit': [45, 60, 54, 66, 57, 63, 50, 53],
    'Quantity': [10, 15, 12, 18, 14, 16, 11, 13]
}

sales_df = pd.DataFrame(sales_data)

groupby = sales_df.groupby('Region')
print("Groupby Region:")
print(groupby.groups)
total_sum =groupby['Sales'].sum()
print("Total Sum:")
print(total_sum)
avg_profit = groupby['Profit'].mean()
print("Average Profit:")
print(avg_profit)
total_quantity = groupby['Quantity'].sum()
print("Total Quantity:")
print(total_quantity)
````

````python
df_groupby_region = {
    'Total_Sales': total_sum,
    'Avg_Profit': avg_profit,
    'Total_Quantity': total_quantity
}
df = pd.DataFrame(df_groupby_region)
df = df.sort_values('Total_Sales', ascending=False)
print(df)
````

````python
best_index = np.argmax(total_sum.values)
best_region = total_sum.index[best_index]
print(f"Top Region: {best_region}")
````

#Multi-level Grouping

````python
employee_data = {
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'Sales', 'IT', 'Sales'],
    'Level': ['Junior', 'Senior', 'Senior', 'Junior', 'Junior', 'Senior', 'Mid', 'Mid'],
    'Salary': [3000, 5000, 5500, 2800, 3200, 6000, 4000, 3500],
    'Age': [25, 35, 32, 24, 27, 38, 29, 31],
    'Experience': [2, 10, 8, 1, 3, 12, 5, 6]
}

employee_df = pd.DataFrame(employee_data)
print(employee_df)
````

````python
goupby_dep_lev = employee_df.groupby(['Department', 'Level'])
print(goupby_dep_lev.groups)
````

````python
avg_salary = goupby_dep_lev['Salary'].mean()
print("Average Salary:")
print(avg_salary)
````

````python
avg_age = goupby_dep_lev['Age'].mean()
print("Average Age:")
print(avg_age)
````

````python
total_employees_count = goupby_dep_lev.size()
print("Total Employees Count:")
print(total_employees_count)
````

````python
dep_lev = {
    'Average Salary': avg_salary,
    'Average Age': avg_age,
    'Total Employees Count': total_employees_count
}

df = pd.DataFrame(dep_lev)
print(df)
````

````python
groupby_dep = employee_df.groupby('Department')
print(groupby_dep.groups)
````

````python
salary_std = groupby_dep['Salary'].apply(np.std)
print(salary_std)
````

````python
min_expereince = groupby_dep['Experience'].min()
print("Minimum experience:")
print(min_expereince)
````

````python
max_expereince = groupby_dep['Experience'].max()
print("Maximum experience:")
print(max_expereince)
````

````python
dep_data = {
    'Salary Standard Deviation': salary_std,
    'Minimum Experience': min_expereince,
    'Maximum Experience': max_expereince
}

df = pd.DataFrame(dep_data)
print(df)
````

#Custom Aggregations

````python
transactions = {
    'Store_ID': ['S1', 'S2', 'S1', 'S3', 'S2', 'S3', 'S1', 'S2', 'S3'],
    'Category': ['Food', 'Electronics', 'Food', 'Clothing',
                 'Electronics', 'Food', 'Clothing', 'Food', 'Electronics'],
    'Amount': [50, 500, 75, 120, 450, 60, 90, 55, 480],
    'Discount': [5, 50, 8, 12, 45, 6, 9, 6, 48],
    'Items': [3, 1, 4, 2, 1, 3, 2, 3, 1]
}

transactions_df = pd.DataFrame(transactions)
print(transactions_df)
````

````python
groupby_id = transactions_df.groupby('Store_ID')
print("Groupby Store ID:")
print(groupby_id.groups)
````

````python
total_revenue = groupby_id['Amount'].sum()
print("Total Revenue:")
print(total_revenue)
````

````python
total_dicount = groupby_id['Discount'].sum()
print("Total Discount:")
print(total_dicount)
````

````python
avg_transactions = groupby_id['Items'].mean()
print("Average Transactions:")
print(avg_transactions)
````

````python
total_item_sold = groupby_id['Items'].sum()
print("Total Item Sold:")
print(total_item_sold)
````

````python
discount_pct = (total_dicount / total_revenue) * 100
avg_items_per_trans = total_item_sold / groupby_id.size()
````

````python
metric = {
    'Total Revenue': total_revenue,
    'Total Discount': total_dicount,
    'Average Transactions': avg_transactions,
    'Total Item Sold': total_item_sold,
    'Discount ptc': discount_pct,
    'Average Items Sold': avg_items_per_trans
}

df = pd.DataFrame(metric)
print(df)
````

````python
max_revenue = groupby_id['Amount'].max()
print("Maximum Revenue:")
print(max_revenue)
````
