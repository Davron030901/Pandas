#Temperature Data Cleaning (30%)

````python
temperatures = pd.Series([23.5, np.nan, 25.1, np.nan, 24.8, 26.3, np.nan, 25.9])
count_nan = temperatures.isna().sum()
print(count_nan)
clean_data = temperatures.dropna()
print(clean_data)
````

````python
forward_fill = temperatures.fillna(method='ffill')
print(forward_fill)
````

````python
print(forward_fill.mean())
````

````python
fill_temperatures = temperatures.fillna(temperatures.mean())
print(fill_temperatures)
````

# Sales Data Interpolation

````python
sales_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [150, np.nan, 180, np.nan, np.nan, 220],
    'Profit': [45, 50, np.nan, 60, 65, np.nan]
}

df = pd.DataFrame(sales_data)
print(df)
````

````python
each_column_nan = df.isna().sum()
print(each_column_nan)
````

````python
df['Sales'] = df['Sales'].fillna(df['Sales'].interpolate(method='linear'))
print(df['Sales'])
````

````python
df['Profit'] = df['Profit'].fillna(df['Profit'].mean())
print(df['Profit'])
````

````python
print(df)
````

#Student Grades Analysis

````python
grades = {
    'Student': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri'],
    'Math': [85, np.nan, 92, 78, np.nan],
    'Physics': [np.nan, 88, 91, np.nan, 86],
    'Chemistry': [80, 85, np.nan, 90, 88]
}

df = pd.DataFrame(grades)
print(df)
````

````python
each_student_missing = df.isna().sum(axis=1)
print(each_student_missing)
````

````python
for i in range(len(df)):
    if each_student_missing[i] >= 2:
        df.drop(i, inplace=True)
print(df)
````

````python
df['Chemistry'] = df['Chemistry'].fillna(df['Chemistry'].mean())
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['Physics'] = df['Physics'].fillna(df['Physics'].mean())
print(df)
````

````python
avg_student = df.mean(numeric_only=True, axis=1)
print(avg_student)
````

````python
add_df = df.assign(Average=avg_student)
print(add_df)
````
