#Transform - Group Statistics to Original Shape

````python
sales = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South', 'North'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, 150, 120, 180, 110]
})
````

````python
print("Original DataFrame:")
print(sales)
````

````python
sales['Region_Avg'] = sales.groupby('Region')['Sales'].transform('mean')

print("\nDataFrame with Region Average:")
print(sales)
````

````python
sales['Diff_form_Avg'] = sales['Sales'] - sales['Region_Avg']

print("\nDataFrame with Difference from Average:")
print(sales)
````

#Apply - Custom Functions

````python
students = pd.DataFrame({
    'Class': ['A', 'A', 'B', 'B', 'A'],
    'Student': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri'],
    'Score': [85, 92, 78, 88, 95]
})

print("Students DataFrame:")
print(students)
````

````python
def score_range(group):
  return group.max() - group.min()

result = students.groupby('Class')['Score'].apply(score_range)

print("\nScore Range by Class:")
print(result)
````

````python
def top_2_students(group):
  return group.nlargest(2, 'Score')

top_students = students.groupby('Class').apply(top_2_students)

print("\nTop 2 Students by Class:")
print(top_students)
````

#Filter - Conditional Group Selection

````python
stores = pd.DataFrame({
    'Store': ['S1', 'S1', 'S2', 'S2', 'S3', 'S3'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Sales': [1000, 1200, 800, 750, 1500, 1600]
})

print('All Stores:')
print(stores)
````

````python
high_performers = stores.groupby('Store').filter(lambda x: x['Sales'].mean() > 1000)

print('\nHigh Performers:')
print(high_performers)
````

#Rank within Groups

````python
employees = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Sales', 'Sales'],
    'Name': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri', 'Zara'],
    'Salary': [5000, 5500, 4000, 4200, 3500, 3800]
})

print("Employees:")
print(employees)
````

````python
employees['Rank'] = employees.groupby('Department')['Salary'].rank(ascending=False)

print("\nEmployees with Ranks:")
print(employees)
````

````python
top_earns = employees[employees['Rank'] == 1]

print("Top Earners by Department:")
print(top_earns)
````

#Sales Performance Analysis

````python
sales_data = {
    'Salesperson': ['Ali', 'Vali', 'Guli', 'Ali', 'Vali', 'Guli', 'Ali', 'Vali'],
    'Region': ['North', 'North', 'South', 'North', 'North', 'South', 'North', 'North'],
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Sales': [5000, 6000, 4500, 5500, 6200, 4800, 5200, 6500]
}

sales_df = pd.DataFrame(sales_data)
print("Sales Data:")
print(sales_df)
````

````python
sales_df["Region_Avg_Sales"] = sales_df.groupby('Region')['Sales'].transform('mean')

print("\nSales Data with Region Average:")
print(sales_df)
````

````python
def performance_category(row):
  if row['Sales'] > row['Region_Avg_Sales']:
    return 'Above'
  else:
    return 'Below'
````

````python
sales_df['Performance'] = sales_df.apply(performance_category, axis=1)

print("\nSales Data with Performance Category:")
print(sales_df)
````

````python
region_avg = sales_df.groupby('Region')['Sales'].mean()
print("Region averages:")
print(region_avg)

high_performing_regions = region_avg[region_avg > 5000].index

filtered_df = sales_df[sales_df['Region'].isin(high_performing_regions)]
print(filtered_df)
````

#Student Ranking System

````python
exam_data = {
    'Student': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri', 'Zara',
                'Ali', 'Vali', 'Guli', 'Sami', 'Nuri', 'Zara'],
    'Subject': ['Math', 'Math', 'Math', 'Math', 'Math', 'Math',
                'Science', 'Science', 'Science', 'Science', 'Science', 'Science'],
    'Score': [85, 92, 78, 88, 95, 80, 90, 88, 92, 85, 87, 94]
}


exam_df = pd.DataFrame(exam_data)
print("Exam Data:")
print(exam_df)
````

````python
rank = exam_df.groupby('Subject')['Score'].rank(ascending=False)

exam_df['Rank'] = rank

print("\nExam Data with Ranks:")
print(exam_df)
````

````python
def score_range(group):
    return group.max() - group.min()
````

````python
result = exam_df.groupby('Student')['Score'].apply(score_range)

print("\nScore Range per Class:")
print(result)
````

````python
subject_median = exam_df.groupby('Subject')['Score'].transform('median')

exam_df['Subject_Median'] = subject_median

print("\nMedian Score per Subject:")
print(exam_df)
````

````python
dif_from_median = exam_df['Score'] - exam_df['Subject_Median']

exam_df['Difference from Subject Median'] = dif_from_median
print(exam_df)
````

````python
rank = exam_df[exam_df['Rank'] <= 3]

print("\nStudents with Highest Scores:")
print(rank)
````

#Product Category Performance

````python
products = {
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Food', 'Food',
                 'Electronics', 'Clothing', 'Food', 'Electronics'],
    'Product_ID': ['E1', 'E2', 'C1', 'C2', 'F1', 'F2', 'E3', 'C3', 'F3', 'E4'],
    'Price': [500, 800, 50, 80, 15, 20, 600, 60, 18, 750],
    'Units_Sold': [20, 15, 100, 80, 200, 180, 18, 90, 190, 16],
    'Rating': [4.5, 4.8, 4.2, 4.0, 4.6, 4.4, 4.7, 4.3, 4.5, 4.6]
}

products_df = pd.DataFrame(products)
print("Products Data:")
print(products_df)
````

````python
revinue = products_df['Price'] * products_df['Units_Sold']

products_df['Revenue'] = revinue
````

````python
def total(series):
  return series.sum()

total_revinue = products_df.groupby('Category')['Revenue'].apply(total)

print("\nProducts Data with Total Revenue:")
print(total_revinue)
````

````python
def average(series):
  return series.mean()

average_rating = products_df.groupby('Category')['Rating'].apply(average)
print("\nProducts Data with Average Rating:")
print(average_rating)
````

````python
def count(series):
  return series.count()

product_count = products_df.groupby('Category')['Product_ID'].apply(count)
print("\nProducts Data with Product Count:")
print(product_count)
````

````python
products_df['Category_Total_Revenue'] = products_df.groupby('Category')['Revenue'].transform('sum')

print("\nProducts Data with Category Total Revenue:")
print(products_df)
````

````python
products_df['Revenue_Share_%'] = products_df['Revenue'] / products_df['Category_Total_Revenue'] * 100

print("\nProducts Data with Revenue Share %:")
print(products_df)
````

````python
high_rated_categories = average_rating[average_rating >= 4.5].index
final_df = products_df[products_df['Category'].isin(high_rated_categories)]

print("High-Rated Categories Products:")
print(final_df)
````

````python
revunue_rank = products_df.groupby('Category')['Revenue'].rank(ascending=False)

products_df['Revenue_Rank'] = revunue_rank
print("\nProducts Data with Revenue Rank:")
print(products_df)
````
