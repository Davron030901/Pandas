**Plotting**
Pandas uses the *plot()* method to create diagrams.

We can use Pyplot, a submodule of the Matplotlib library to visualize the diagram on the screen.

Read more about Matplotlib in our Matplotlib Tutorial.

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import IPython.display as display
df = pd.read_csv('/content/data.csv')

df.plot()
plt.savefig('grafik.png')
plt.show()
plt.close()  # Plotni tozalash (ochish) plt.show() bilan aralashmaslik uchun

# Faylni yuklab, ekranda ko'rsatish
img = Image.open('grafik.png')
display.display(img)
```

**Scatter Plot**
Specify that you want a scatter plot with the kind argument:

*kind = 'scatter'*

A scatter plot needs an x- and a y-axis.

In the example below we will use "Duration" for the x-axis and "Calories" for the y-axis.

Include the x and y arguments like this:

*x = 'Duration', y = 'Calories'*

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import IPython.display as display
df = pd.read_csv('data.csv')

df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')

print(df.head())
plt.show()
plt.savefig('grafic.png')

# Grafikni ekranda ko'rsatish
plt.close()  # Plotni tozalash (ochish) plt.show() bilan aralashmaslik uchun

# Faylni yuklab, ekranda ko'rsatish
img = Image.open('grafic.png')
display.display(img)
```

***Histogram***
Use the kind argument to specify that you want a histogram:

*kind = 'hist'*

A histogram needs only one column.

A histogram shows us the frequency of each interval, e.g. how many workouts lasted between 50 and 60 minutes?

In the example below we will use the "Duration" column to create the histogram:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

df["Duration"].plot(kind = 'hist')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import IPython.display as display

# CSV faylni o'qing
df = pd.read_csv('/content/data.csv')

# "Duration" ustuni bo'yicha histogram chizish
df["Duration"].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')  # bins bilan ustunlar sonini boshqaramiz

# y o'qining balandligini cheklash
plt.ylim(0, 50)  # y o'qini 0 dan 50 gacha cheklaymiz, kerak bo'lsa bu qiymatni moslang

# Grafikni ekranda ko'rsatish
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.title("Histogram of Duration")
plt.show()

# Grafikni fayl sifatida saqlash
plt.savefig('graficing.png')

# Plotni tozalash
plt.close()

# Faylni yuklab, ekranda ko'rsatish
imge = Image.open('graficing.png')
display.display(imge)
```
