DataFrame to Array

````python
df = pd.DataFrame({'A':[1, 2, 3], 'B': [4, 5, 6]})
arr = df.values
print("Df array:", arr)
````

````python
arr = df.to_numpy()
print("Df array:", arr)
````

Array to DataFrame

````python
arr = np.array([[1, 2], [3, 4], [5, 6]])
df = pd.DataFrame(arr, columns=['X', 'Y'])
print("Df array:")
print(df)
````

#CSV → NumPy → Statistics

````python
data = {
    'Temperature': [23.5, 25.1, 24.8, 26.3, 25.9],
    'Humidity': [65, 70, 68, 72, 71],
    'Pressure': [1013, 1015, 1014, 1016, 1015]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)
````

````python
arr = df.values
print("Array:")
print(arr)
````

````python
arr = df.to_numpy()
print("Array:")
print(arr)
````

````python

mean = arr.mean(axis=0)
print("Mean:")
print(mean)
````

````python
std = arr.std(axis=0)
print("Standard Deviation:")
print(std.shape)
````

````python
data = {
    'Mean': mean,
    'Std': std
}
````

````python
df = pd.DataFrame(data, index=['Temperature', 'Humidity', 'Pressure'])
print("DataFrame:")
print(df)
````

# Array Manipulation → DataFrame

````python
array = np.random.randint(0, 100, size=(4, 3))
print("Array:")
print(array)

min = array.min(axis=1)
max = array.max(axis=1)
mean = array.mean(axis=1)
data = {
    'Row': [0, 1, 2, 3],
    'Min': min,
    'Max': max,
    'Mean': mean
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
````

#Real Dataset Processing

````python
sales_data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South'],
    'Sales': [150, 200, 175, 160, 190, 185],
    'Profit': [45, 60, 52, 48, 57, 55]
}

print("Sales Data:")
df = pd.DataFrame(sales_data)
print(df)
````

````python
convert_num = df.to_numpy()
print("Convert Num:")
print(convert_num)
````

````python
sales = np.array(convert_num[:, 2])
profit = np.array(convert_num[:, 3])

print("Sales:", sales)
print("Profit:", profit)
````

````python
profit_percentage = (profit / sales) * 100
print("Profit Percentage:")
print(profit_percentage)
````

````python
add_df = pd.DataFrame(profit_percentage, columns=['Profit Margin'])
print("DataFrame:")
print(add_df)
````

````python
all_df = pd.concat([df, add_df], axis=1)
print("DataFrame:")
print(all_df)
````

````python
region = np.array(all_df['Region'])
print("Region:", region)
````

````python
regiaon_north = (region == 'North')
print("Region North:", regiaon_north)
regiaon_south = (region == 'South')
print("Region South:", regiaon_south)
````

````python
regiaon_north_profit = profit_percentage[regiaon_north]
print("Region North Profit margin:", regiaon_north_profit)
regiaon_south_profit = profit_percentage[regiaon_south]
print("Region South Profit margin:", regiaon_south_profit)
region
````

````python
mean_north_profit = regiaon_north_profit.mean()
print("Mean Region North Profit:", mean_north_profit)
mean_south_profit = regiaon_south_profit.mean()
print("Mean Region South Profit:", mean_south_profit)
````
