```python
import numpy as np
import pandas as pd
```

#E-commerce Analytics

```python
sales_data = {
    'Product_ID': ['P001', 'P002', 'P003', 'P004', 'P005'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Food', 'Clothing'],
    'Price': [599.99, 49.99, 899.99, 15.50, 79.99],
    'Quantity_Sold': [45, 120, 30, 200, 85],
    'Rating': [4.5, 4.2, 4.8, 3.9, 4.6]
}

df = pd.DataFrame(sales_data)
print("DataFrame:")
print(df)
```

```python
numeric_to_numpy = df[['Price', 'Quantity_Sold', 'Rating']].to_numpy()
print("Numpy Array:")
print(numeric_to_numpy)
```

```python
revenue = numeric_to_numpy[:, 0] * numeric_to_numpy[:, 1]
print("Revenue Per Product:")
print(revenue)
```

```python
total_revenue = revenue.sum()
avg_revenue = revenue.mean()
print(f"Total Revenue: ${total_revenue:.2f}")
print(f"Average Revenue: ${avg_revenue:.2f}")
```

```python
max_index = revenue.argmax()
max_product = df.loc[max_index, 'Product_ID']
print("Most Profitable Product:")
print(max_product)
```

```python
df_add_revenue = df.copy()
df_add_revenue['Revenue'] = revenue
print("DataFrame with Revenue:")
print(df_add_revenue)
```

#Weather Station Data Cleanup

```python
weather_data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temp_C': [23.5, np.nan, 25.8, np.nan, 24.2, 26.1, np.nan],
    'Humidity': [65, 70, np.nan, 72, np.nan, 68, 71],
    'Wind_Speed': [12, np.nan, 15, np.nan, np.nan, 14, 13],
    'Pressure': [1013, 1015, 1014, np.nan, 1016, np.nan, 1015]
}

df = pd.DataFrame(weather_data)
print("DataFrame:")
print(df)
```

```python
missing_values_count = df.isnull().sum()
print("Missing Values Count:")
print(missing_values_count)
```

```python
df['Temp_C'] = df['Temp_C'].interpolate(method='linear')
df['Humidity'] = df['Humidity'].fillna(method='ffill')
df['Wind_Speed'] = df['Wind_Speed'].mean()
df['Pressure'] = df['Pressure'].fillna(method='bfill')
print("DataFrame with Linear Interpolation:")
print(df)
```

```python
df_to_numpy_num = df.iloc[:, 1:].to_numpy()
print("Numpy Array:")
print(df_to_numpy_num)
```

```python
mean = np.mean(df_to_numpy_num, axis=0)
min = np.min(df_to_numpy_num, axis=0)
max = np.max(df_to_numpy_num, axis=0)
std = np.std(df_to_numpy_num, axis=0)
summary = {
    'Matric' : ['Temp_C', 'Humidity', 'Wind_Speed', 'Pressure'],
    'Mean' : mean,
    'Min' : min,
    'Max' : max,
    'Std' : std
    }

df_summary = pd.DataFrame(summary)
print("Summary Statistics:")
print(df_summary)
```

#Student Performance Analysis

```python
student_scores = {
    'Name': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri', 'Zara'],
    'Math': [85, np.nan, 92, 78, np.nan, 88],
    'English': [np.nan, 88, 85, np.nan, 90, 87],
    'Science': [90, 85, np.nan, 88, np.nan, np.nan],
    'History': [82, np.nan, 89, 85, 91, np.nan]
}

df = pd.DataFrame(student_scores)
print("DataFrame:")
print(df)
```

```python
missing_values_count = df.isnull().sum()
print("Missing Values Count:")
print(missing_values_count)
```

```python
missing_per_student = df.iloc[:, 1:].isnull().sum(axis=1)
students_with_3plus = (missing_per_student >= 3).sum()
print("Count of Students with More Than 3 Missing Scores:")
print(students_with_3plus)
```

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['English'] = df['English'].fillna(df['English'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['History'] = df['History'].fillna(df['History'].mean())
print("DataFrame with Mean Imputation:")
print(df)
```

```python
df_to_num = df.iloc[:, 1:].to_numpy()
print("Numpy Array:")
print(df_to_num)
```

```python
avrg = np.mean(df_to_num, axis=1)
grade = np.where(avrg >= 90, 'A', np.where(avrg >= 80, 'B', np.where(avrg >= 70, 'C', np.where(avrg >= 60, 'D', 'F'))))
print("Average Scores:")
print(avrg)
print("Grades:")
print(grade)
#
```

```python
df["Average"] = avrg
print("DataFrame with Average Scores:")
print(df)
```

```python
df["Grade"] = grade
print("DataFrame with Grades:")
print(df)
```
