#E-commerce Analytics

```python
sales_data = {
    'Product_ID': ['P001', 'P002', 'P003', 'P004', 'P005'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Food', 'Clothing'],
    'Price': [599.99, 49.99, 899.99, 15.50, 79.99],
    'Quantity_Sold': [45, 120, 30, 200, 85],
    'Rating': [4.5, 4.2, 4.8, 3.9, 4.6]
}

df = pd.DataFrame(sales_data)
print("DataFrame:")
print(df)
```

```python
numeric_to_numpy = df.iloc[:, 2:].to_numpy()
print("Numpy Array:")
print(numeric_to_numpy)
```

```python
revenue = numeric_to_numpy[:, 0] * numeric_to_numpy[:, 1]
print("Revenue Per Product:")
print(revenue)
```
